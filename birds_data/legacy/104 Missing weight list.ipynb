```python
# import pandas to deal with DataFrame
import pandas as pd
# import os lib to check if a file is exist
import os
```

```python
# load dataframe file
birds_df = pd.read_csv("birds_April_13.csv")
# reset index
birds_df.set_index('Idx', inplace=True)
```

```python
birds_df.head()
```

```python
missing_length_idx = birds_df[birds_df['max_weight'].isnull()].index
```

```python
# check how many birds still need length
len(missing_length_idx)
```

```python
# create a new dataframe to store the birds missing images
missing_birds_df = birds_df.ix[missing_length_idx, ['Common name', 'Binomial', 'min_weight', 'max_weight']]
```

```python
# check the result
missing_birds_df
```

```python
# load dataframe file
new_df = pd.read_csv("104 missing weight list.csv")
new_df.set_index("Idx", inplace=True)
columns = ['min_weight', 'max_weight']
```

```python
for idx in missing_length_idx:
    for column in columns:
        birds_df.ix[idx, column] = new_df.ix[idx, column]
```

```python
birds_df.to_csv("birds_April_13.csv")
```

```python
# save into a new csv
missing_birds_df.to_csv("104 missing weight list.csv")
```
