```python
vic_df = pd.read_html(
    "http://avibase.bsc-eoc.org/checklist.jsp?lang=EN&p2=1&list=avibase&synlang=&region=AUvi&version=text&lifelist=&highlight=0")
vic_df = vic_df[1].copy()
vic_df = vic_df[15:-1]
vic_df = vic_df[[0,1,3]]
vic_df.columns = ['common_name', 'binomial', 'status']
vic_df = vic_df[vic_df['binomial'].notnull()]
```

```python
sa_df = pd.read_html(
    "http://avibase.bsc-eoc.org/checklist.jsp?lang=EN&p2=1&list=avibase&synlang=&region=AUsa&version=text&lifelist=&highlight=0")
sa_df = sa_df[1].copy()
sa_df = sa_df[15:-1]
sa_df = sa_df[[0,1,3]]
sa_df.columns = ['common_name', 'binomial', 'status']
sa_df = sa_df[sa_df['binomial'].notnull()]
```

```python
wa_df = pd.read_html(
    "http://avibase.bsc-eoc.org/checklist.jsp?lang=EN&p2=1&list=avibase&synlang=&region=AUwa&version=text&lifelist=&highlight=0")
wa_df = wa_df[1].copy()
wa_df = wa_df[15:-1]
wa_df = wa_df[[0,1,3]]
wa_df.columns = ['common_name', 'binomial', 'status']
wa_df = wa_df[wa_df['binomial'].notnull()]
```

```python
location_str = ['vic', 'nsw', 'qld', 'act', 'nt', 'sa', 'tas', 'wa']
location_df = [vic_df, nsw_df, qld_df, act_df, nt_df, sa_df, tas_df, wa_df]
```

```python
for state in location_str:
    birds_df[state] = pd.np.nan
```

```python
for state_idx in range(len(location_str)):
    state_df = location_df[state_idx]
    state_str = location_str[state_idx]
    for idx in birds_df.index:
        if birds_df.ix[idx, 'Binomial'] in state_df.binomial.values:
            birds_df.ix[idx, state_str] = 1
```
